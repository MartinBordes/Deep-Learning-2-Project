# 1 - Variation du nombre de couches

```python
## Paramètres

p_small = [320, 320]
p_mid = [320, 170, 100, 50]
p_big = [320, 120, 80, 60, 40, 20]

learning_rate = 0.01
len_batch = 10
n_epochs = 1000

nb_images = 16
nb_iter = 200
size_img = (20, 16)

numbers = [4, 5, 6, 7, 8, 9]
X = lire_alpha_digits(path = "binaryalphadigs.mat", numbers = numbers)
```

```python
dbn_mid = DBN(p_mid)
dbn_mid.train(X, learning_rate=learning_rate, len_batch=len_batch, n_epochs=n_epochs)
images = dbn_mid.generer_image(nb_images=nb_images, nb_iter=nb_iter, size_img=size_img)
plt.savefig('figures/alpha_couches_mid.jpg')
plt.show()
```

```python
dbn_big = DBN(p_big)
dbn_big.train(X, learning_rate=learning_rate, len_batch=len_batch, n_epochs=n_epochs)
images = dbn_big.generer_image(nb_images=nb_images, nb_iter=nb_iter, size_img=size_img)
plt.savefig('figures/alpha_couches_big.jpg')
plt.show()
```
